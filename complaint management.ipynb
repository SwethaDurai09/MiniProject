```python
import mysql.connector
from datetime import datetime
import tkinter as tk
from tkinter import messagebox

# Connect to the MySQL database
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="complaint"
)

cursor = db.cursor()

# Function to register a complaint
def register_complaint():
    customer_name = customer_name_entry.get()
    complaint_description = complaint_description_entry.get()
    
    # Get the current date and time
    current_time = datetime.now()
    
    # Insert the complaint into the database
    insert_query = "INSERT INTO complaints (customer_name, complaint_description, status, created_at) VALUES (%s, %s, 'Open', %s)"
    data = (customer_name, complaint_description, current_time)
    cursor.execute(insert_query, data)
    db.commit()
    messagebox.showinfo("Success", "Complaint registered successfully.")

# Function to view and update complaints
def view_and_update_complaints():
    # View all open complaints
    view_query = "SELECT * FROM complaints WHERE status = 'Open'"
    cursor.execute(view_query)
    complaints = cursor.fetchall()

    if not complaints:
        messagebox.showinfo("Info", "No open complaints.")
        return

    open_complaints_text = "Open Complaints:\n"
    for complaint in complaints:
        open_complaints_text += f"Complaint ID: {complaint[0]}, Customer Name: {complaint[1]}, Description: {complaint[2]}\n"

    open_complaints_label.config(text=open_complaints_text)

    complaint_id = int(complaint_id_entry.get())

    # Update the status of the selected complaint
    update_status = update_status_entry.get().capitalize()
    update_query = "UPDATE complaints SET status = %s WHERE id = %s"
    data = (update_status, complaint_id)
    cursor.execute(update_query, data)
    db.commit()
    messagebox.showinfo("Success", "Complaint status updated successfully.")
    view_and_update_complaints()  # Refresh the displayed complaints

# Create the main window
root = tk.Tk()
root.title("Complaint Management System")

# Create labels and entry widgets for complaint registration
tk.Label(root, text="Customer Name:").pack()
customer_name_entry = tk.Entry(root)
customer_name_entry.pack()

tk.Label(root, text="Complaint Description:").pack()
complaint_description_entry = tk.Entry(root)
complaint_description_entry.pack()

# Create a button for registering a complaint
register_button = tk.Button(root, text="Register Complaint", command=register_complaint)
register_button.pack()

# Create labels and entry widgets for viewing and updating complaints
tk.Label(root, text="Complaint ID to Update:").pack()
complaint_id_entry = tk.Entry(root)
complaint_id_entry.pack()

tk.Label(root, text="New Status (Open/Closed):").pack()
update_status_entry = tk.Entry(root)
update_status_entry.pack()

# Create a button for viewing and updating complaints
view_update_button = tk.Button(root, text="View and Update Complaints", command=view_and_update_complaints)
view_update_button.pack()

# Create a label to display open complaints
open_complaints_label = tk.Label(root, text="", wraplength=300)
open_complaints_label.pack()

# Main loop for the Tkinter GUI
root.mainloop()

# Close the database connection when done
db.close()
```
